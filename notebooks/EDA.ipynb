```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the processed dataset
data = pd.read_csv(r'C:\Users\91952\Desktop\Credit card fraud detection\Data\credit_card_data_processed.csv')

# Display the first few rows
print(data.head())

# Summary statistics
print(data.describe())

# Check for class distribution
sns.countplot(x='Class', data=data)
plt.title('Class Distribution')
plt.show()

# Correlation matrix
corr_matrix = data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# tells which features are correlated.
corr_pairs = corr_matrix.unstack()
sorted_pairs = corr_pairs.sort_values(kind="quicksort", ascending=False)
strong_pairs = sorted_pairs[sorted_pairs != 1]  # Remove self-correlations (correlation of 1)
print("Most correlated features:\n", strong_pairs[0:15])

fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Plot for Amount
sns.histplot(data['Amount'], bins=50, kde=True, ax=ax[0])
ax[0].set_title('Distribution of Amount')
ax[0].set_xlabel('Amount')
ax[0].set_ylabel('Frequency')

# Plot for Time
sns.histplot(data['Time'], bins=50, kde=True, ax=ax[1])
ax[1].set_title('Distribution of Time')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Frequency')

plt.show()

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the processed dataset
data = pd.read_csv(r'C:\Users\91952\Desktop\Credit card fraud detection\Data\credit_card_data_processed.csv')

# Function to summarize distribution
def summarize_distribution(feature_name):
    feature_data = data[feature_name]
    summary = {
        "mean": np.mean(feature_data),
        "median": np.median(feature_data),
        "std_dev": np.std(feature_data),
        "min": np.min(feature_data),
        "max": np.max(feature_data),
        "skewness": feature_data.skew(),
        "kurtosis": feature_data.kurt()
    }
    return summary

# Summarize distributions for 'Time' and 'Amount'
time_summary = summarize_distribution('Time')
amount_summary = summarize_distribution('Amount')

# Print summaries
print("Distribution Summary for 'Time':")
print(f"Mean: {time_summary['mean']}")
print(f"Median: {time_summary['median']}")
print(f"Standard Deviation: {time_summary['std_dev']}")
print(f"Min: {time_summary['min']}")
print(f"Max: {time_summary['max']}")
print(f"Skewness: {time_summary['skewness']}")
print(f"Kurtosis: {time_summary['kurtosis']}")

print("\nDistribution Summary for 'Amount':")
print(f"Mean: {amount_summary['mean']}")
print(f"Median: {amount_summary['median']}")
print(f"Standard Deviation: {amount_summary['std_dev']}")
print(f"Min: {amount_summary['min']}")
print(f"Max: {amount_summary['max']}")
print(f"Skewness: {amount_summary['skewness']}")
print(f"Kurtosis: {amount_summary['kurtosis']}")
```
